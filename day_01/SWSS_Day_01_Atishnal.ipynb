```python
from math import factorial
from math import pi

def cos_approx(x, accuracy =10):
    """ The cosine function approximation using its taylor expansion at accuracy level 10 """
    
    ans = [] # make empty list
    for n in range(accuracy):
        term = (((-1)**n)/factorial(2*n))* (x**(2*n))
        ans.append(term)
        
    approx = sum(ans) # sums all the values
    print(ans)    
    return approx
        
cos_approx(pi)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def spherical_cartesian(radius, azimuth, zentith):
    """ 
    Converts spherical coordinates to cartesian 
    azimuth & zentith should be in radian
    
    """
    x = radius*np.sin(zentith)*np.cos(azimuth)
    y = radius*np.sin(zentith)*np.cos(azimuth)
    z = radius*np.cos(zentith)
    return x, y, z

fig = plt.figure(dpi=60) # better control
fig.set_size_inches(10,10)
axes = plt.axes(projection = '3d') # make 3d axes
r = np. linspace(0,1) # domain of r from 0 to 1
theta = np.linspace(0,2*np.pi) # provides theta values in radians
phi = np.linspace(0,2*np.pi) # phi values in radians
x,y,z = spherical_cartesian(r, theta, phi)
axes.plot(x,y,z)
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_zlabel('z')

Test1 = spherical_cartesian(1,0,0)
numpy.allclose(a, b, rtol=1e-05, atol=1e-08, equal_nan=False)
```
